```python
import pandas as pd
import json
import os
os.environ["PATH"] += os.pathsep + r"C:\\Program Files\\Graphviz\bin"

from sklearn.model_selection import train_test_split
from matplotlib import pyplot as plt
from sklearn import datasets
from sklearn.tree import DecisionTreeClassifier 
from sklearn import tree
from dtreeviz import dtreeviz
from sklearn.tree import export_graphviz
import subprocess
from sklearn.preprocessing import LabelEncoder
```

```python
columns = [
    "name",
    "level",
    "perception",
    "str",
    "dex",
    "con",
    "int",
    "wis",
    "cha",
    "ac",
    "hp",
    "fort",
    "refl",
    "will"
]
df = pd.DataFrame(columns = columns)

directory = os.fsencode("bestiary")
i = 0
for file in os.listdir(directory):
    filename = os.fsdecode(file)
    with open(f"bestiary\\{filename}", 'r') as f:
        try:
            data = json.load(f)
        except UnicodeDecodeError:
            continue

    line = []

    line.append(data["name"])
    line.append(str(data["system"]["details"]["level"]["value"]))
    line.append(data["system"]["perception"]["mod"])
    line.append(data["system"]["abilities"]["str"]["mod"])
    line.append(data["system"]["abilities"]["dex"]["mod"])
    line.append(data["system"]["abilities"]["con"]["mod"])
    line.append(data["system"]["abilities"]["int"]["mod"])
    line.append(data["system"]["abilities"]["wis"]["mod"])
    line.append(data["system"]["abilities"]["cha"]["mod"])
    line.append(data["system"]["attributes"]["ac"]["value"])
    line.append(data["system"]["attributes"]["hp"]["max"])
    line.append(data["system"]["saves"]["fortitude"]["value"])
    line.append(data["system"]["saves"]["reflex"]["value"])
    line.append(data["system"]["saves"]["will"]["value"])
    
    df.loc[i] = line
    i += 1

df

```

```python
X = df.drop(columns=["name", "level"])
y = df["level"]

# X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 101)

clf = DecisionTreeClassifier()
model = clf.fit(X, y)
```

```python
list_o_nums = [i for i in range(-2,25)]

export_graphviz(
    clf, 
    feature_names=X.columns,  
    class_names=list_o_nums,
    filled=True, rounded=True, node_ids=True, out_file='tree.dot'
)

# Convert DOT to SVG
subprocess.call(['dot', '-Tsvg', 'tree.dot', '-o', 'tree.svg'])
```
